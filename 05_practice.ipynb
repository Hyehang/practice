```python
mid = mid.rename(columns = {'poptotal':'total',
           'popasian':'asian'})
mid
```

```python
mid['a/t'] = (mid['asian']/mid['total'])*100
mid
```

```python
mid['a/t'].plot.hist()
```

```python
mid['a/t'].mean()
```

```python
mid['asian_p'] = np.where(mid['a/t']> mid['a/t'].mean(),'large','small')
```

```python
mid
```

```python
count_asian = mid['asian_p'].value_counts()
count_asian
```

```python
count_asian.plot.bar(rot=0)
```
