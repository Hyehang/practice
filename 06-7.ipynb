## 분석 도전하기

#### Q1. popadults는 해당 지역의 성인 인구, poptotal은 전체 인구를 나타냅니다. midwest 데이터에 '전체 인구 대비 미성년 인구 백분율' 변수를 추가하세요.

```python
midwest = pd.read_csv('midwest.csv')
```

```python
midwest.head()
```

```python
midwest = midwest.assign(na_p = lambda x : ((x['poptotal']-x['popadults'])/x['poptotal'])*100)
midwest.head()
```

### Q2.미성년 인구 백분율이 가장 높은 상위 5개 country의 미성년 인구 백분율을 출력하세요.

```python
midwest.sort_values('na_p',ascending = False) \
        .head() \
        [['county','na_p']]
```

#### Q3. 분류표의 기준에 따라 미성년 비율 등급 변수를 추가하고. 각 등급에 몇개의 지역이 있는지 알아보세요.

```python
import numpy as np
```

```python
midwest = midwest.assign( level = np.where(midwest['na_p']>= 40,'large', np.where(midwest['na_p']>=30,'middle','small'))) 
    
```

```python
midwest['level'].value_counts()
```

```python
midwest.groupby('level')\
        .agg(n = ('level','count'))
```

#### Q4. popasian은 해당 지역의 아시아인 인구를 나타냅니다. ' 전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고 하위 10개 지역의 state(주),county(지역),아시아인 인구 백분율을 출력하세요.

```python
midwest.assign(a_ratio = lambda x : (x['popasian']/x['poptotal'])*100) \
        .sort_values('a_ratio') \
        .head(10) \
        [['state','county','a_ratio']]
```
