### 텍스트 마이닝 : 문자로 된 데이터에서 가치 있는 정보를 얻어 내는 분석 기법

```python
moon = open('speech_moon.txt',encoding = 'UTF-8').read()
```

```python
moon
```

```python
import re
```

```python
moon = re.sub('[^가-힣]',' ',moon) #한글이 아닌 모든 문자를 공백으로 만듬
```

```python
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
```

```python
import pandas as pd
```

```python
df_word = pd.DataFrame({'word' : nouns})
```

```python
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
```

```python
df_word = df_word.query('count >= 2')
```

```python
df_word.sort_values('count')
```

```python
df_word = df_word.groupby('word', as_index = False )\
                .agg(n=('word','count')) \
                .sort_values('n' , ascending = False )
```

```python
df_word
```

```python
top20 = df_word.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.family' : 'Malgun Gothic',
                    'figure.dpi' : '120',
                    'figure.figsize' : [6.5,6]})
```

```python
sns.barplot(data = top20 , y = 'word' , x = 'n')
```

### 워드 클라우드 

```python
font = "DoHyeon-Regular.ttf"
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,    #난수 고정
               font_path = "DoHyeon-Regular.ttf",   #폰트 설정
               width = 400 ,      #가로크기
               height = 400 ,      #세로크기
               background_color = 'white')   #배경색
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

### 워드 클라우드 모양 바꾸기

```python
import PIL
icon = PIL.Image.open('cloud.png')
```

```python
import numpy as np
img = PIL.Image.new('RGB',icon.size,(255,255,255))
img.paste(icon,icon)
img = np.array(img)
```

```python
wc = WordCloud( random_state = 1234,
              font_path = font,
              width = 400,
              height = 400,
              background_color = 'white',
              mask = img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

#### 기사 댓글 가져오기

```python
df['reply'] = df['reply'].str.replace('[^가-힣]',' ',regex = True )
df['reply'].head()
```

```python
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
nouns = df['reply'].apply(kkma.nouns)
```

```python
nouns
```

```python
nouns = nouns.explode()
```

```python
nouns
```

```python
df_word = pd.DataFrame({'word' : nouns})
```

```python
df_word['count'] = df_word['word'].str.len()
```

```python
df_word = df_word.query('count >= 2')
```

```python
df_word
```

```python
df_word = df_word.groupby('word',as_index=False) \
                .agg( n = ('word','count')) \
                .sort_values('n' ,ascending = False)
```
